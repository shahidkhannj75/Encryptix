This dataset is highly unbalanced


0-------->Normal
1-------->Fraud

```python
#Seperating the data for analysis
legit=df[df.Class==0]
fraud=df[df.Class==1]
print(legit.shape,fraud.shape)
```

Statistical measures of the data

```python
#Compare the values for both transactions
df.groupby('Class').mean()
```

Undersampling

Build a sample dataset containing similar distribution of normal transaction and Fraudlent transaction

```python
legit_sample=legit.sample(n=492)
```

Concatinating 2 DataFrames

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Spltting the data into Features and Target

```python
x=new_dataset.drop(columns='Class',axis=1)
y=new_dataset['Class']
```

```python
x
```

```python
y
```

Split Data into Training and Testing Data

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model Training

```python
model=LogisticRegression()
```

```python
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

```python
#Accuracy on training data
x_train_prediction=model.predict(x_train)
training_data_accuracy=accuracy_score(x_train_prediction,y_train)
print("Training data accuracy",training_data_accuracy)
report=classification_report(x_train_prediction,y_train)
report
```

```python
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
print("Testing data Accuracy",test_data_accuracy)
report=classification_report(x_test_prediction,y_test)
report
```
